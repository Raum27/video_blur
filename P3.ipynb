```python
from numpy import dot
from numpy.linalg import norm
import cv2
import cv2,mmcv
import time
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
```

```python
def cosine_similarity(a, b):
    a = np.array(a,dtype=np.float32)
    b = np.array(b,dtype=np.float32)
    value_AB = dot(a, b) / (norm(a) * norm(b))
    return 0.0 if value_AB != value_AB else value_AB
```

```python
video = mmcv.VideoReader(r"C:\Users\Raum\Desktop\jec\code\dataface\videoplayback.mp4")
print('Frames: ',len(video))

MASK_FRAMES = np.zeros(len(video),dtype=np.int32)

pos_frame = 1

# ''' pull image from folder '''
frame_maker = cv2.resize(cv2.cvtColor(video[0],cv2.COLOR_BGR2RGB),(224,224)).reshape(-1)

for i in range(len(video)):
  frame = cv2.resize(cv2.cvtColor(video[i],cv2.COLOR_BGR2RGB),(224,224)).reshape(-1)
  if cosine_similarity(frame_maker,frame)== 0.0:
    MASK_FRAMES[i]=0
    frame_maker = cv2.resize(cv2.cvtColor(video[i],cv2.COLOR_BGR2RGB),(224,224)).reshape(-1)
    continue
  elif cosine_similarity(frame_maker,frame)>= 0.7:
    MASK_FRAMES[i]=pos_frame
  else :
    frame_maker = cv2.resize(cv2.cvtColor(video[i],cv2.COLOR_BGR2RGB),(224,224)).reshape(-1)
    pos_frame +=1
    MASK_FRAMES[i]=pos_frame
print(MASK_FRAMES.shape)
print(MASK_FRAMES[55:66])
```
